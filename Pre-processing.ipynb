```
import os
import random

def remove_random_images(base_folder_path, subfolder, num_to_remove):
    subfolder_path = os.path.join(base_folder_path, subfolder)
    image_files = os.listdir(subfolder_path)
    
    if len(image_files) < num_to_remove:
        print(f"Tidak cukup file gambar dalam subfolder {subfolder} untuk dihapus.")
        return
    
    files_to_remove = random.sample(image_files, num_to_remove)
    
    for file in files_to_remove:
        file_path = os.path.join(subfolder_path, file)
        os.remove(file_path)
        print(f"File {file} berhasil dihapus.")

# Contoh penggunaan
subfolder = 'papula_pustula'
num_to_remove = 20

remove_random_images(base_folder_path, subfolder, num_to_remove)
```

### Pengecekan File Duplikasi

```
def find_duplicate_files(folder_path):
    seen_files = set()
    duplicates = []

    for subdir, _, files in os.walk(folder_path):
        for file in files:
            file_path = os.path.join(subdir, file)
            # Extracting file name
            file_name = os.path.basename(file_path)
            if file_name in seen_files:
                duplicates.append(file_path)
            else:
                seen_files.add(file_name)

    if duplicates:
        print(f"Found {len(duplicates)} duplicate files based on file names:")
        for dup in duplicates:
            print(dup)
    else:
        print("No duplicate files found.")
```

```
import os
from PIL import Image

def convert_to_jpg(base_folder_path):
    # List of subfolders
    subfolders = ['acne_nodules', 'whitehead', 'blackhead', 'milia', 'papula_pustula']

    # Loop through each subfolder
    for subfolder in subfolders:
        folder_path = os.path.join(base_folder_path, subfolder)

        # Process each file in the subfolder
        for filename in os.listdir(folder_path):
            if filename.endswith(('.png', '.jpeg', '.jfif')):
                try:
                    image_path = os.path.join(folder_path, filename)
                    img = Image.open(image_path)

                    # Save the converted image with the same filename
                    new_filename = os.path.splitext(filename)[0] + '.jpg'
                    new_image_path = os.path.join(folder_path, new_filename)
                    img = img.convert('RGB')
                    img.save(image_path, 'JPEG')

                    # Remove the original file extension
                    os.rename(image_path, new_image_path)

                    print(f"Converted {filename} to {new_filename} in {subfolder}")
                except Exception as e:
                    print(f"Failed to process {filename} in {subfolder}: {e}")

# Base folder path
base_folder_path = '/work/20240520-092144/DATASET_CLEANING/Training/'

# Call the function
convert_to_jpg(base_folder_path)
```

### Mengkonversi Warna Citra ke RGB dan Merescale Pixel Intensitas Warna Gambar Untuk Memiliki Rentang 0 -1

```
def normalize_images(base_folder_path, target_size):
    subfolders = ['acne_nodules', 'whitehead', 'blackhead', 'papula_pustula', 'milia']
    for subfolder in subfolders:
        input_folder_path = os.path.join(base_folder_path, subfolder)
        for filename in os.listdir(input_folder_path):
            if filename.lower().endswith(('.jpg')):
                try:
                    image_path = os.path.join(input_folder_path, filename)
                    img = Image.open(image_path).convert('RGB')  # Ensure image is in RGB mode
                    # Normalize the image to [0, 1] range
                    img_array = np.array(img) / 255.0
                    # Convert back to PIL Image
                    normalized_img = Image.fromarray((img_array * 255).astype('uint8'))
                    new_image_path = os.path.join(input_folder_path, filename)
                    normalized_img.save(new_image_path)
                    print(f"Normalized {filename} and saved to {input_folder_path}")
                except Exception as e:
                    print(f"Failed to process {filename} in {subfolder}: {e}")

base_folder_path = '/work/20240520-092144/DATASET_CLEANING/Training_resize/'
normalize_images(base_folder_path, target_size)
```

### Augmentasi data dengan transformasi shear (pergeseran pada setiap titik dalam gambar)

```
import os
import cv2
import numpy as np

# Direktori input dan output
base_folder_path = '/work/20240520-092144/DATASET_CLEANING/Training_resize/'
output_base_folder = '/work/20240520-092144/DATASET_CLEANING/Training_shear/'

# Subfolders
subfolders = ['acne_nodules', 'whitehead', 'blackhead', 'papula_pustula', 'milia']

# Definisikan fungsi untuk membuat direktori jika belum ada
def create_dir(path):
    if not os.path.exists(path):
        os.makedirs(path)

# Fungsi untuk menerapkan transformasi shear
def apply_shear(image, shear_value=0.2):
    rows, cols, _ = image.shape
    M = np.float32([[1, shear_value, 0], [0, 1, 0]])
    sheared_img = cv2.warpAffine(image, M, (int(cols + rows * shear_value), rows))
    return sheared_img

# Loop melalui setiap subfolder
for subfolder in subfolders:
    input_folder_path = os.path.join(base_folder_path, subfolder)
    output_folder_path = os.path.join(output_base_folder, subfolder)
    create_dir(output_folder_path)  # Membuat direktori output jika belum ada

    # Loop melalui setiap file gambar dalam subfolder
    for filename in os.listdir(input_folder_path):
        if filename.lower().endswith(('.jpg', '.png')):
            try:
                image_path = os.path.join(input_folder_path, filename)
                img = cv2.imread(image_path)

                # Augmentasi dengan transformasi shear
                sheared_img = apply_shear(img, shear_value=0.2)
                output_filename = f"aug_shear_{os.path.splitext(filename)[0]}.jpg"
                output_path = os.path.join(output_folder_path, output_filename)
                cv2.imwrite(output_path, sheared_img)
                print(f"Saved augmented image (shear): {output_path}")

            except Exception as e:
                print(f"Failed to process {filename} in {subfolder}: {e}")

print("Augmentasi transformasi shear selesai.")
```

### Augmentasi data dengan shifting (pergeseran) secara horizontal dan vertikal

```
import os
import cv2
import numpy as np

# Direktori input dan output
base_folder_path = '/work/20240520-092144/DATASET_CLEANING/Training_resize/'
output_base_folder = '/work/20240520-092144/DATASET_CLEANING/Training_shift/'

# Subfolders
subfolders = ['acne_nodules', 'whitehead', 'blackhead', 'papula_pustula', 'milia']

# Definisikan fungsi untuk membuat direktori jika belum ada
def create_dir(path):
    if not os.path.exists(path):
        os.makedirs(path)

# Fungsi untuk menerapkan shifting
def apply_shift(image, x_shift, y_shift):
    rows, cols, _ = image.shape
    M = np.float32([[1, 0, x_shift], [0, 1, y_shift]])
    shifted_img = cv2.warpAffine(image, M, (cols, rows))
    return shifted_img

# Loop melalui setiap subfolder
for subfolder in subfolders:
    input_folder_path = os.path.join(base_folder_path, subfolder)
    output_folder_path = os.path.join(output_base_folder, subfolder)
    create_dir(output_folder_path)  # Membuat direktori output jika belum ada

    # Loop melalui setiap file gambar dalam subfolder
    for filename in os.listdir(input_folder_path):
        if filename.lower().endswith(('.jpg', '.png')):
            try:
                image_path = os.path.join(input_folder_path, filename)
                img = cv2.imread(image_path)
                rows, cols, _ = img.shape

                # Menentukan besar pergeseran (shifting)
                x_shift = np.random.uniform(-0.2 * cols, 0.2 * cols)
                y_shift = np.random.uniform(-0.2 * rows, 0.2 * rows)

                # Augmentasi dengan shifting
                shifted_img = apply_shift(img, x_shift, y_shift)
                x_shift_percent = int(x_shift / cols * 100)
                y_shift_percent = int(y_shift / rows * 100)
                output_filename = f"aug_shift_x{x_shift_percent}y{y_shift_percent}_{os.path.splitext(filename)[0]}.jpg"
                output_path = os.path.join(output_folder_path, output_filename)
                cv2.imwrite(output_path, shifted_img)
                print(f"Saved augmented image (shift): {output_path}")

            except Exception as e:
                print(f"Failed to process {filename} in {subfolder}: {e}")

print("Augmentasi shifting selesai.")
```

### Copy semua hasil augmentasi pada 1 folder

```
import os
import shutil

def copy_files_by_type(src_base_dir, dst_base_dir, subfolders):
    """
    Copies files from source directories to destination directory based on subfolder types.

    :param src_base_dir: The base directory where the source files are located.
    :param dst_base_dir: The base directory where the files will be copied to.
    :param subfolders: List of subfolder names that specify types of files.
    """
    # Loop through each subfolder
    for subfolder in subfolders:
        src_dir = os.path.join(src_base_dir, subfolder)
        dst_dir = os.path.join(dst_base_dir, subfolder)
        
        # Create the destination directory if it does not exist
        if not os.path.exists(dst_dir):
            os.makedirs(dst_dir)
            print(f"Created directory: {dst_dir}")
        
        # Loop through each file in the source directory
        for filename in os.listdir(src_dir):
            src_path = os.path.join(src_dir, filename)
            dst_path = os.path.join(dst_dir, filename)
            
            # Copy file from source to destination
            shutil.copy2(src_path, dst_path)
            print(f"Copied {filename} from {src_dir} to {dst_dir}")

    print("File copying completed for this batch.")
```
